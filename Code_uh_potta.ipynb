```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load the CSV file (change the path as needed)
file_path = 'object_small.csv'  # Make sure the file is in the same directory
df = pd.read_csv(file_path)
```

```python

# Rename columns to x, y, z
df.columns = ['x', 'y', 'z']
```

```python

# Calculate dimensions
width = df['x'].max() - df['x'].min()
height = df['y'].max() - df['y'].min()
depth = df['z'].max() - df['z'].min()

print("üìè Object Dimensions:")
print(f"Width (x-axis):  {width:.2f} units")
print(f"Height (y-axis): {height:.2f} units")
print(f"Depth (z-axis):  {depth:.2f} units")
```

```python

# Plot the 3D point cloud
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['x'], df['y'], df['z'], c='blue', s=5)
```

```python

# Labels and title
ax.set_xlabel('X (Width)')
ax.set_ylabel('Y (Height)')
ax.set_zlabel('Z (Depth)')
ax.set_title('3D Point Cloud of Object')

plt.show()
```
